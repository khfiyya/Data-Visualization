# Data Visualization

```python
###Who were the passengers on the titanic? (What age, gender, class etc)

###Gender Plot
sns.factorplot('Sex',data=df_titanic,kind='count')

### Shows more male passengers than female 
```

Shows more male passengers than female 

```python
### Class plot
sns.factorplot('Pclass',data=df_titanic,kind='count')
```

Chances of survival, Pclass 3

```python
 sns.factorplot('Pclass',data=df_titanic,hue='Sex',kind='count')
```

More passengers are from class Three, and female so high.

```python
sns.catplot(x = "Pclass", y="Survived", data = df_titanic, kind="bar", height = 6)
```

Chances of survival : 1st Class > 2nd Class > 3rd Class

```python
df_titanic['Siblings/Spouses Aboard'].unique()
sns.catplot(x = "Siblings/Spouses Aboard", y = "Survived", data = df_titanic, kind="bar", height = 8)
```

From the Bar Graph we can say that Passengers with 1 or 2 siblings have a better chance of survival than those have more than 2 siblings.

```python
sns.FacetGrid(df_titanic, col="Survived", height = 7).map(sns.distplot, "Age").set_ylabels("Survival Probability")
```

From the distribution plot we can say that people with more age have a lesser chance of survival than people with less age

```python
sns.barplot(x="Sex", y="Survived", data=df_titanic)
```

From above it is crystal clear that "There were more Female survivors than compared to Male survivors

```python
m = sns.lmplot(x = 'Age', y = 'Fare', data = df_titanic, hue = 'Sex', fit_reg=False)


lm.set(title = 'Fare & Age')


axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

And same , who have paid higher faresThere were more Female survivors than compared to Male survivors and hat people with more age have a lesser chance of survival than people with less age.

```python
sns.heatmap(data=df_titanic.corr(), annot=True)
plt.show()
```

From the heatmap we can say that those who have paid higher fares have a better chance of Survival
